```python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

## Importar Dataset

```python
df = pd.read_csv('/content/Temp_test.csv', sep=';')
df.set_index('Name', inplace=True)
display(df.head())
```

### Especies y UORA

```python
bacillus_amyloliquefaciens_treatment = df.loc['B. amyloliquefaciens']
bacillus_amyloliquefaciens_control = df.loc['Control1']

bacillus_subtilis_treatment = df.loc['B. subtilis']
bacillus_subtilis_control = df.loc['Control2']

bacillus_licheniformis_treatment = df.loc['B. licheniformis']
bacillus_licheniformis_control = df.loc['Control3']

uora_treatment = df.loc['UORA']
uora_control = df.loc['Control4']
```

```python
# Calculate standard deviation for B. amyloliquefaciens
std_amyloliquefaciens_treatment = np.std(bacillus_amyloliquefaciens_treatment)
std_amyloliquefaciens_control = np.std(bacillus_amyloliquefaciens_control)

print(f"B. amyloliquefaciens - Desviación Estándar (Tratamiento): {std_amyloliquefaciens_treatment:.4f}")
print(f"B. amyloliquefaciens - Desviación Estándar (Control): {std_amyloliquefaciens_control:.4f}")
print("-" * 20)

# Calculate standard deviation for B. subtilis
std_subtilis_treatment = np.std(bacillus_subtilis_treatment)
std_subtilis_control = np.std(bacillus_subtilis_control)

print(f"B. subtilis - Desviación Estándar (Tratamiento): {std_subtilis_treatment:.4f}")
print(f"B. subtilis - Desviación Estándar (Control): {std_subtilis_control:.4f}")
print("-" * 20)

# Calculate standard deviation for B. licheniformis
std_licheniformis_treatment = np.std(bacillus_licheniformis_treatment)
std_licheniformis_control = np.std(bacillus_licheniformis_control)

print(f"B. licheniformis - Desviación Estándar (Tratamiento): {std_licheniformis_treatment:.4f}")
print(f"B. licheniformis - Desviación Estándar (Control): {std_licheniformis_control:.4f}")
```

```python
# Shapiro-Wilk test for B. amyloliquefaciens
shapiro_amyloliquefaciens_treatment = stats.shapiro(bacillus_amyloliquefaciens_treatment)
shapiro_amyloliquefaciens_control = stats.shapiro(bacillus_amyloliquefaciens_control)

print("Shapiro-Wilk Test for Normality:")
print(f"B. amyloliquefaciens (Tratamiento): Statistic={shapiro_amyloliquefaciens_treatment.statistic:.4f}, P-value={shapiro_amyloliquefaciens_treatment.pvalue:.4f}")
print(f"B. amyloliquefaciens (Control): Statistic={shapiro_amyloliquefaciens_control.statistic:.4f}, P-value={shapiro_amyloliquefaciens_control.pvalue:.4f}")
print("-" * 20)

# Shapiro-Wilk test for B. subtilis
shapiro_subtilis_treatment = stats.shapiro(bacillus_subtilis_treatment)
shapiro_subtilis_control = stats.shapiro(bacillus_subtilis_control)

print(f"B. subtilis (Tratamiento): Statistic={shapiro_subtilis_treatment.statistic:.4f}, P-value={shapiro_subtilis_treatment.pvalue:.4f}")
print(f"B. subtilis (Control): Statistic={shapiro_subtilis_control.statistic:.4f}, P-value={shapiro_subtilis_control.pvalue:.4f}")
print("-" * 20)

# Shapiro-Wilk test for B. licheniformis
shapiro_licheniformis_treatment = stats.shapiro(bacillus_licheniformis_treatment)
shapiro_licheniformis_control = stats.shapiro(bacillus_licheniformis_control)

print(f"B. licheniformis (Tratamiento): Statistic={shapiro_licheniformis_treatment.statistic:.4f}, P-value={shapiro_licheniformis_treatment.pvalue:.4f}")
print(f"B. licheniformis (Control): Statistic={shapiro_licheniformis_control.statistic:.4f}, P-value={shapiro_licheniformis_control.pvalue:.4f}")
print("-" * 20)

# Shapiro-Wilk test for UORA
shapiro_uora_treatment = stats.shapiro(uora_treatment)
shapiro_uora_control = stats.shapiro(uora_control)

print(f"UORA (Tratamiento): Statistic={shapiro_uora_treatment.statistic:.4f}, P-value={shapiro_uora_treatment.pvalue:.4f}")
print(f"UORA (Control): Statistic={shapiro_uora_control.statistic:.4f}, P-value={shapiro_uora_control.pvalue:.4f}")
```

```python
# Levene test for B. amyloliquefaciens
levene_amyloliquefaciens = stats.levene(bacillus_amyloliquefaciens_treatment, bacillus_amyloliquefaciens_control)

print("Levene's Test for Homogeneity of Variances:")
print(f"B. amyloliquefaciens: Statistic={levene_amyloliquefaciens.statistic:.4f}, P-value={levene_amyloliquefaciens.pvalue:.4f}")
print("-" * 20)

# Levene test for B. subtilis
levene_subtilis = stats.levene(bacillus_subtilis_treatment, bacillus_subtilis_control)

print(f"B. subtilis: Statistic={levene_subtilis.statistic:.4f}, P-value={levene_subtilis.pvalue:.4f}")
print("-" * 20)

# Levene test for B. licheniformis
levene_licheniformis = stats.levene(bacillus_licheniformis_treatment, bacillus_licheniformis_control)

print(f"B. licheniformis: Statistic={levene_licheniformis.statistic:.4f}, P-value={levene_licheniformis.pvalue:.4f}")
print("-" * 20)

#Levene test for UORA
levene_uora = stats.levene(uora_treatment, uora_control)

print(f"UORA: Statistic={levene_uora.statistic:.4f}, P-value={levene_uora.pvalue:.4f}")
```

```python
# Perform independent samples T-test for B. amyloliquefaciens (assuming equal variances)
ttest_amyloliquefaciens = stats.ttest_ind(bacillus_amyloliquefaciens_treatment, bacillus_amyloliquefaciens_control, equal_var=True)

print("Independent Samples T-test Results:")
print(f"B. amyloliquefaciens: Statistic={ttest_amyloliquefaciens.statistic:.4f}, P-value={ttest_amyloliquefaciens.pvalue:.4f}")
print("-" * 20)

# Perform independent samples T-test for B. subtilis (assuming equal variances)
ttest_subtilis = stats.ttest_ind(bacillus_subtilis_treatment, bacillus_subtilis_control, equal_var=True)

print(f"B. subtilis: Statistic={ttest_subtilis.statistic:.4f}, P-value={ttest_subtilis.pvalue:.4f}")
print("-" * 20)

# Perform independent samples T-test for B. licheniformis (assuming unequal variances - Welch's T-test)
ttest_licheniformis = stats.ttest_ind(bacillus_licheniformis_treatment, bacillus_licheniformis_control, equal_var=False)

print(f"B. licheniformis (Welch's T-test): Statistic={ttest_licheniformis.statistic:.4f}, P-value={ttest_licheniformis.pvalue:.4f}")
print("-" * 20)

# Perform independent samples T-test for UORA (assuming equal variances)
ttest_uora = stats.ttest_ind(uora_treatment, uora_control, equal_var=True)

print(f"UORA: Statistic={ttest_uora.statistic:.4f}, P-value={ttest_uora.pvalue:.4f}")
```

```python
# Gather the results
results_data = {
    "Species": ["B. amyloliquefaciens", "B. subtilis", "B. licheniformis", "UORA"],
    "Shapiro-Wilk (Treatment) Statistic": [shapiro_amyloliquefaciens_treatment.statistic, shapiro_subtilis_treatment.statistic, shapiro_licheniformis_treatment.statistic, shapiro_uora_treatment.statistic],
    "Shapiro-Wilk (Treatment) P-value": [shapiro_amyloliquefaciens_treatment.pvalue, shapiro_subtilis_treatment.pvalue, shapiro_licheniformis_treatment.pvalue, shapiro_uora_treatment.pvalue],
    "Shapiro-Wilk (Control) Statistic": [shapiro_amyloliquefaciens_control.statistic, shapiro_subtilis_control.statistic, shapiro_licheniformis_control.statistic, shapiro_uora_control.statistic],
    "Shapiro-Wilk (Control) P-value": [shapiro_amyloliquefaciens_control.pvalue, shapiro_subtilis_control.pvalue, shapiro_licheniformis_control.pvalue, shapiro_uora_control.pvalue],
    "Levene Statistic": [levene_amyloliquefaciens.statistic, levene_subtilis.statistic, levene_licheniformis.statistic, levene_uora.statistic],
    "Levene P-value": [levene_amyloliquefaciens.pvalue, levene_subtilis.pvalue, levene_licheniformis.pvalue, levene_uora.pvalue],
    "T-test Statistic": [ttest_amyloliquefaciens.statistic, ttest_subtilis.statistic, ttest_licheniformis.statistic, ttest_uora.statistic],
    "T-test P-value": [ttest_amyloliquefaciens.pvalue, ttest_subtilis.pvalue, ttest_licheniformis.pvalue, ttest_uora.pvalue],
    "T-test Type": ["Independent (Equal Var)", "Independent (Equal Var)", "Welch's (Unequal Var)", "Independent (Equal Var)"]
}

results_df = pd.DataFrame(results_data)

# Format P-values for better readability
results_df["Shapiro-Wilk (Treatment) P-value"] = results_df["Shapiro-Wilk (Treatment) P-value"].map('{:.4f}'.format)
results_df["Shapiro-Wilk (Control) P-value"] = results_df["Shapiro-Wilk (Control) P-value"].map('{:.4f}'.format)
results_df["Levene P-value"] = results_df["Levene P-value"].map('{:.4f}'.format)
results_df["T-test P-value"] = results_df["T-test P-value"].map('{:.4f}'.format)


display(results_df)
```

```python
# Calculate the effect of the treatment (difference between control and treatment)
effect_amyloliquefaciens = bacillus_amyloliquefaciens_control - bacillus_amyloliquefaciens_treatment
effect_subtilis = bacillus_subtilis_control - bacillus_subtilis_treatment
effect_licheniformis = bacillus_licheniformis_control - bacillus_licheniformis_treatment
effect_uora = uora_control - uora_treatment

# Perform one-way ANOVA on the effects
anova_effect = stats.f_oneway(effect_amyloliquefaciens, effect_subtilis, effect_licheniformis, effect_uora)

print("One-way ANOVA on the Effect of Temperature:")
print(f"Statistic (F-value): {anova_effect.statistic:.4f}")
print(f"P-value: {anova_effect.pvalue:.4f}")
```

```python
# Combine the effect data into a single DataFrame
effect_values = np.concatenate([effect_amyloliquefaciens, effect_subtilis, effect_licheniformis, effect_uora])
group_labels = (['B. amyloliquefaciens'] * len(effect_amyloliquefaciens) +
                ['B. subtilis'] * len(effect_subtilis) +
                ['B. licheniformis'] * len(effect_licheniformis) +
                ['UORA'] * len(effect_uora))

effect_df = pd.DataFrame({'effect': effect_values, 'group': group_labels})

# Perform Tukey's HSD post-hoc test
tukey_result = pairwise_tukeyhsd(endog=effect_df['effect'], groups=effect_df['group'], alpha=0.05)

print("Tukey's Honestly Significant Difference (HSD) Test:")
print(tukey_result)
```

```python
from statsmodels.stats.power import TTestIndPower
from scipy import stats # Import stats for t.interval
import pandas as pd # Import pandas

# Import the dataset and set the index
df = pd.read_csv('/content/Temp_test.csv', sep=';')
df.set_index('Name', inplace=True)

# Define the variables again to ensure they are available
bacillus_amyloliquefaciens_treatment = df.loc['B. amyloliquefaciens']
bacillus_amyloliquefaciens_control = df.loc['Control1']

bacillus_subtilis_treatment = df.loc['B. subtilis']
bacillus_subtilis_control = df.loc['Control2']

bacillus_licheniformis_treatment = df.loc['B. licheniformis']
bacillus_licheniformis_control = df.loc['Control3']

uora_treatment = df.loc['UORA']
uora_control = df.loc['Control4']


# Function to calculate effect size (Cohen's d), confidence interval, and power
def analyze_group_effect(treatment_data, control_data, alpha=0.05, desired_power=0.80):
    """
    Calculates Cohen's d effect size, confidence interval for the difference in means,
    and statistical power for an independent samples t-test.

    Args:
        treatment_data (pd.Series): Data for the treatment group.
        control_data (pd.Series): Data for the control group.
        alpha (float): Significance level (default is 0.05).
        desired_power (float): Desired statistical power for sample size calculation (default is 0.80).

    Returns:
        dict: A dictionary containing the calculated effect size, confidence interval, and power.
    """
    # Calculate Cohen's d
    n1, n2 = len(treatment_data), len(control_data)
    mean1, mean2 = np.mean(treatment_data), np.mean(control_data)
    s1, s2 = np.std(treatment_data, ddof=1), np.std(control_data, ddof=1) # Use ddof=1 for sample standard deviation

    # Pooled standard deviation for Cohen's d and equal variance CI
    pooled_std = np.sqrt(((n1 - 1) * s1**2 + (n2 - 1) * s2**2) / (n1 + n2 - 2))
    cohen_d = (mean1 - mean2) / pooled_std

    # Calculate confidence interval for the difference in means
    # Using scipy.stats.t.interval
    if treatment_data.name == 'B. licheniformis':
        # Welch's t-test for unequal variances
        diff_mean = mean1 - mean2
        se_diff = np.sqrt(s1**2/n1 + s2**2/n2)
        # Degrees of freedom for Welch's t-test
        dof = (s1**2/n1 + s2**2/n2)**2 / ((s1**2/n1)**2 / (n1 - 1) + (s2**2/n2)**2 / (n2 - 1))
        ci_lower, ci_upper = stats.t.interval(1 - alpha, dof, loc=diff_mean, scale=se_diff)
    else:
        # Independent samples t-test assuming equal variances
        diff_mean = mean1 - mean2
        se_diff = pooled_std * np.sqrt(1/n1 + 1/n2)
        dof = n1 + n2 - 2
        ci_lower, ci_upper = stats.t.interval(1 - alpha, dof, loc=diff_mean, scale=se_diff)


    # Calculate statistical power
    # For the given sample size and calculated effect size
    analysis = TTestIndPower()
    power = analysis.solve_power(cohen_d, nobs1=n1, alpha=alpha, ratio=n2/n1, alternative='two-sided')

    return {
        "Cohen's d": cohen_d,
        "Confidence Interval Lower": ci_lower,
        "Confidence Interval Upper": ci_upper,
        "Statistical Power": power
    }

# Analyze each group
analysis_results = {}

analysis_results['B. amyloliquefaciens'] = analyze_group_effect(bacillus_amyloliquefaciens_treatment, bacillus_amyloliquefaciens_control)
analysis_results['B. subtilis'] = analyze_group_effect(bacillus_subtilis_treatment, bacillus_subtilis_control)
# Note: For B. licheniformis, the CI calculation in the function now uses Welch's method for unequal variances.
analysis_results['B. licheniformis'] = analyze_group_effect(bacillus_licheniformis_treatment, bacillus_licheniformis_control)
analysis_results['UORA'] = analyze_group_effect(uora_treatment, uora_control)

# Print the results
print("Analysis Results (Effect Size, Confidence Interval, and Power):")
for group, results in analysis_results.items():
    print(f"\n{group}:")
    print(f"  Cohen's d: {results['Cohen\'s d']:.4f}")
    print(f"  Confidence Interval (95%): ({results['Confidence Interval Lower']:.4f}, {results['Confidence Interval Upper']:.4f})")
    print(f"  Statistical Power: {results['Statistical Power']:.4f}")
```

**Interpretation of Results:**

*   **Cohen's d:** Represents the effect size. A common interpretation is:
    *   0.2: Small effect
    *   0.5: Medium effect
    *   0.8: Large effect
    The sign of Cohen's d indicates the direction of the effect (positive if the treatment mean is higher than the control mean, negative if lower).

*   **Confidence Interval:** Provides a range of plausible values for the true difference in means between the treatment and control groups. If the interval includes zero, it suggests that the observed difference is not statistically significant at the chosen alpha level.

*   **Statistical Power:** Represents the probability of finding a statistically significant result if a true effect of the calculated size exists. A power of 0.80 or higher is generally considered good. Low power might indicate that the sample size is too small to detect a true effect.

Keep in mind that the confidence interval calculation for *B. licheniformis* in this code assumes equal variances, which was contradicted by the Levene's test results where unequal variances were indicated. For a more precise confidence interval for *B. licheniformis*, a method specifically designed for unequal variances would be needed.
